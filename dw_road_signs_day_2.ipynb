```
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```
train = pd.read_pickle('train.p')
signs = pd.read_csv('signnames.csv')
```

```
train.keys()
```

```
x_train, y_train = train['features'], train['labels']
x_train.shape, y_train.shape
```

```
plt.imshow(x_train[0])
plt.title(y_train[0]);
```

```
signs.head()
```

```
dict_signs = signs.to_dict()['b']
dict_signs
```

```
x_train[y_train == 3].shape
```

```
for id_sign, desc in dict_signs.items():
  given_signs = x_train[y_train == id_sign]
  plt.figure(figsize=(20,10))
  for i in range(4):
    plt.subplot(1,4, i+1)
    plt.imshow(given_signs[i])
    plt.axis('off')
    plt.title(desc)
  plt.tight_layout()
  plt.show()
```

```
cnt = Counter(y_train).most_common()

id_labels, cnt_labels = zip(*cnt)
ids = range(len(id_labels), 0, -1)

plt.figure(figsize=(20,10))
labels = [dict_signs[id_] for id_ in id_labels]
plt.barh(ids, cnt_labels);
plt.xticks()
```

```
cnt = Counter(y_train).most_common()

id_labels, cnt_labels = zip(*cnt)
ids = range(len(id_labels))

plt.figure(figsize=(20,10))
labels = [dict_signs[id_labels[id_]] for id_ in id_labels]
plt.bar(ids, cnt_labels)
plt.xlabel('Signs')
plt.xticks(ids, labels, rotation = 'vertical')
plt.show()
```
